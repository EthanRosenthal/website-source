```python
import os
import requests
import urllib.request
```

```python
def get_thumb(mid):
    try:
        response = requests.get('https://sketchfab.com/i/models/{}'.format(mid)).json()
        thumb = [x['url'] for x in response['thumbnails']['images'] 
                 if (x['width'] == 200 and x['height']==200)
                     or (x['width'] == 256)]
        return thumb[0]
    except:
        print(f'{mid} thumb is missing')
        return None

def save_mid(mid, path):
    thumb = get_thumb(mid)
    if thumb is not None:
        filename = os.path.join(path, f'{mid}.jpg')
        print(filename)
        urllib.request.urlretrieve(thumb, filename)
```
